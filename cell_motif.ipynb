```r
library(dplyr)
library(pheatmap)
library("ggcorrplot")
library(stringr)
library("ggsci")
library("ggplot2")
library("gridExtra")
```

```r
b<-c("LNcap","K562","293T")
all.motif1<-data.frame()
for (name in b) {
  data<-read.table(paste0(name,"/","knownResults.txt"),sep="\t",header=T,row.names = NULL)%>%filter(P.value <= 0.01)
  data$Motif.Name<-as.character(data$Motif.Name)
  data<-data[c(1:15),]
  data_motif<-as.data.frame(data$Motif.Name)
  colnames(data_motif)[1]<-"Motif.Name"
  all.motif1<-rbind(data_motif,all.motif1)
}
all.motif1 <- base::unique(all.motif1)
dim(all.motif1)
```

```r
A<-as.data.frame(all.motif1)
colnames(A)<-"Motif"
head(A)
```

```r
A
```

## LNcap 

```r
LNcap<-read.table("LNcap/knownResults.txt",sep="\t",header=T)
LNcap$Type<-"LNcap"
colnames(LNcap)[1]<-"Motif"
LNcap<-left_join(A,LNcap,by="Motif")
head(LNcap)
LNcap <- LNcap %>%
  mutate(b = as.numeric(gsub("%", "", b)),
         d = as.numeric(gsub("%", "", d)))
LNcap$FC<-as.numeric(LNcap$b)/as.numeric(LNcap$d)


LNcap<-LNcap[,c(1,4,10,11)]
LNcap$`-Log.P.value`<-abs(LNcap$Log.P.value)
#LNcap<-na.omit(LNcap)
LNcap$Motif<-word(LNcap$Motif,1,sep = fixed('(')) 
LNcap$Motif<-toupper(LNcap$Motif)
LNcap$`-Log.P.value`<-scale(LNcap$`-Log.P.value`)
LNcap$FC<-scale(LNcap$FC)
head(LNcap)
```
